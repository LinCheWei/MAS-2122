```python
# import statements
import math
import matplotlib.pyplot as plt
```

```python
# grid generation with nestedloop

# create an empty figure
plt.figure(figsize = (15,15))
# plt.axis('equal')
plt.axis('off')

# generate a 2D grid made of points
N = 25
x,y = [],[]
d = 10.0

for i in range(N):
    for j in range(N):
        x.append(i*d)
        y.append(j*d)

plt.scatter(x, y, s=100, c='red')
```

```python
# diagonals version

plt.figure(figsize = (15,15))
# plt.axis('equal')
plt.axis('off')

N=12
x,y = [],[]
d = 1.0

for i in range(N):
    x.append(i*d)
    y.append(i*d)
plt.scatter(x, y, s=900, c='green')

for i in range(N):
    x[i] += 1.0
plt.scatter(x, y, s=600, c='red')

for i in range(N):
    y[i] += 2.0
plt.scatter(x, y, s=100, color='blue')
```

```python
# square version

plt.figure(figsize = (15,15))
plt.axis('equal')
plt.axis('off')

Nx=14
Ny=6
d = 1.0

x1, y1 = [],[]
x2, y2 = [],[]
x3, y3 = [],[]
x4, y4 = [],[]

for i in range(Nx):
    x1.append(i*d)
    y1.append(0)
plt.scatter(x1,y1, s=1000, c='red')

for i in range(Nx):
    x2.append(i*d)
    y2.append((Ny-1)*d)
plt.scatter(x2, y2, s=1000, c='green') 

for j in range(Ny):
    x3.append((Nx-1)*d)
    y3.append(j*d)
plt.scatter(x3, y3, s=400, c='blue') 

for j in range(Ny):
    x4.append(0)
    y4.append(j*d)
plt.scatter(x4, y4, s=400, c='k') 
```

# Functions

```python
from random import random as rand

def draw_circle(N, radius, center_x, center_y):
    x,y,c = [],[],[]
    
    for i in range(N):
        theta = math.pi * 2 * i/N
        x.append(radius * math.cos(theta) + center_x)
        y.append(radius * math.sin(theta) + center_y)
        c.append([rand()*abs(math.cos(theta)), rand(),  rand()*abs(math.sin(theta)), rand()])
    
    plt.scatter(x,y,s=50*radius, c=c) 
```

```python
plt.figure(figsize = (15,15))
plt.axis('off')

draw_circle(N=30, radius=50, center_x=0, center_y=0)
```

```python
from random import random as rand

plt.figure(figsize = (15,15))
plt.axis('equal')
plt.axis('off')

for i in range(30):
    draw_circle(N=20, 
                radius=rand()*18, 
                center_x=rand()*100, 
                center_y=rand()*100)
```
